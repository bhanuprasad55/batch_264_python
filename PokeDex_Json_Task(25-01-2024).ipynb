```python
fobj = open(r"C:\Users\gali bhanu prasad\Downloads\pokedex.json")
jlist = json.load(fobj)
df = pd.json_normalize(jlist, record_path=['pokemon'])
df = df.explode('weaknesses') # exploding the 'weakness' column to creat rows for each weakness 
df.head()
```

```python
def get_nums(x):
    try: #try to creat iterater from x
        iterator = iter(x)
    except TypeError:   #if x is not iterable, return none
        return None
    else:  #if x iterable extract 'num' field from each element and return a list
        return [c['num'] for c in x]
fobj = open(r"C:\Users\gali bhanu prasad\Downloads\pokedex.json")
jlist = json.load(fobj)
df = pd.json_normalize(jlist, record_path=['pokemon'])

df['evolutions'] = df['next_evolution'].apply( lambda x: get_nums(x) )
df['ancestors'] = df['prev_evolution'].apply( lambda x: get_nums(x) )
# extracting 'weaknesses','evolutions','ancestors' into seperate column using pd.series
weaknesses = df['weaknesses'].apply( pd.Series )
evolutions = df['evolutions'].apply( pd.Series )
ancestors = df['ancestors'].apply( pd.Series )
weaknesses.head()
```

```python
# Rename'weaknesses','evolutions','ancestors' columns with a prefix for clarity and consistency
weaknesses = weaknesses.rename(columns = lambda x: 'wk_' + str(x))
evolutions = evolutions.rename(columns = lambda x: 'ev_' + str(x))
ancestors = ancestors.rename(columns = lambda x: 'an_' + str(x))
weaknesses.head()
```

```python
df = df.merge(weaknesses, left_index=True, right_index=True).merge(evolutions, left_index=True, right_index=True).merge(ancestors, left_index=True, right_index=True)
df.head()  ## Merge the original DataFrame 'df' with the 'weaknesses' DataFrame based on index.includes columns from both DataFrames and making sure for appropate for data structure
```

```python
evens = df.iloc[lambda x: x.index % 2 == 0]# Select rows from the DataFrame 'df' where the index is even
# The lambda function x.index % 2 == 0 generates a boolean mask for even indices
evens.head()
```

```python
#1:3 selects rows with integer indices 1 and 2 (exclusive of 3).
#0:3 selects columns with integer indices 0, 1, and 2.
df.iloc[1:3, 0:3]
```

```python
# Select rows where 'candy_count' is greater than 30
# and columns 'name', 'candy_count', and 'candy'
slice_candies = df.loc[lambda df: df['candy_count'] > 30, ['name', 'candy_count', 'candy']]
slice_candies.head()
```

```python
df.loc[lambda df: np.isnan(df['candy_count']), ['candy_count']] = 0
df.head() ##converting NaN['Not a number' ]values into '0' in 'candy_count' column.
```
